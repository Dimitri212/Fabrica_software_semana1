# 2 - Agrupar o banco de dados a partir da raça/etnia dos alunos, e plotar um gráfico com a nota média em Matemática para cada grupo disponível

```python
#Novo dataframe com race/ethnicity x nota média de matemática

df2 = df.groupby('race/ethnicity')[['math score']].mean()
df2
```

###Gráfico de raça/etnia dos alunos x nota média em Matemática:

```python
df2.plot(kind = 'bar')
plt.title("raça/etnia dos alunos x nota média em Matemática")
```



# 3 - Agrupe o banco de dados a partir do Nível de Escolaridade dos pais, e plote um gráfico com a nota média em Escrita para cada grupo disponível;

```python
#Novo dataframe com Nível de Escolaridade dos pais x nota média em Escrita

df3 = df.groupby('parental level of education')[['writing score']].mean()
df3
```

###Gráfico de Nível de Escolaridade dos pais x nota média em Escrita:

```python
df3.plot(kind = 'bar')
plt.title("Nível de Escolaridade dos pais x nota média em Escrita ")
```

# 4 - Agrupar o banco de dados a partir da coluna referente à participação em um curso preparatório para os testes, e plotar um gráfico com a nota média em Leitura para cada grupo disponível

```python
#Novo dataframe com a participação em um curso preparatório para os testes x nota média em Leitura

df4 = df.groupby('test preparation course')[['reading score']].mean()
df4
```

###Gráfico de participação em um curso preparatório para os testes x nota média em Leitura:

```python
df4.plot(kind = 'bar')
plt.title("curso preparatório para os testes x nota média em Leitura")
```
